```python
tup=(20,10,16,19,25,1,276,188)
print(tup)
for temp in tup:
    if ((temp%2)==1):
        print(temp, "is odd")
        
```

```python
tup2=(20,10,16,19,25,1,276,188)
print(tup2)
for temp2 in tup2:
    if((temp2%2)==0):
        print(temp2, "is even")
        
```
