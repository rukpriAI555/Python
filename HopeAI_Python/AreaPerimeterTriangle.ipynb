```python
def AreaPeriTriangle():
    ht=int(input("Height: "))
    br=int(input("Breadth: "))
    Area=(ht*br)/2
    print("Area of Triangle: ",Area)
    ht1=int(input("Height1: "))
    ht2=int(input("Height2: "))
    br=int(input("Breadth: "))
    Perimeter=ht1+ht2+br
    print("Perimeter of Triangle: ", Perimeter)
```

```python
AreaPeriTriangle()
```
